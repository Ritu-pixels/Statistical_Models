We are conducting a program while using Hidden Markov model where a student is observed with 3 actions: Study, Play or movie. The term hidden refers to the first order Markov process behind the observation. Observation refers to the data we know and can observe. Here the Hidden states are: Free_time and Busy_time

```python
states = ('Free_time', 'Busy_time')
 
observations = ('study', 'play', 'movie')
 
start_probability = {'Free_time': 0.6, 'Busy_time': 0.4}
 
transition_probability = {
   'Free_time' : {'Free_time': 0.7, 'Busy_time': 0.3},
   'Busy_time' : {'Free_time': 0.4, 'Busy_time': 0.6},
   }
 
emission_probability = {
   'Free_time' : {'study': 0.1, 'play': 0.4, 'movie': 0.5},
   'Busy_time' : {'study': 0.6, 'play': 0.3, 'movie': 0.1},
   }

```

```python
from hmmlearn import hmm
import numpy as np

model = hmm.MultinomialHMM(n_components=2)
model.startprob_ = np.array([0.6, 0.4])
model.transmat_ = np.array([[0.7, 0.3],
                            [0.4, 0.6]])
model.emissionprob_ = np.array([[0.1, 0.4, 0.5],
                                [0.6, 0.3, 0.1]])
```

Problem 1, Given a known model what is the likelihood of sequence O happening? i.e Study obeservation

```python
import math

math.exp(model.score(np.array([[0]])))
# 0.30000000000000004
math.exp(model.score(np.array([[1]])))
# 0.36000000000000004
math.exp(model.score(np.array([[2]])))
# 0.3400000000000001
math.exp(model.score(np.array([[2,2,2]])))
# 0.04590400000000001
```
