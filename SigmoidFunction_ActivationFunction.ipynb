```python
import numpy as np
import matplotlib.pyplot as plt 
import random


```

```python
#Sigmoid function:
def sigma(x,w,b):
    return 1/(1+np.exp(-((w*x)+b)))
```

```python
#this is to check how slope (w in wx+b) with different values influnces the sigmoid function 
for w in range(10):
    X = np.linspace(-1,1,100)
    plt.plot(X,sigma(X,w,0))
    plt.text(-0.5,0.5,'w='+str(w)) #to show w value on the graphs
    plt.show()
```
